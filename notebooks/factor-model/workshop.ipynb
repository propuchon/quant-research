```python
import pandas as pd
import numpy as np

import yfinance as yf
```

```python
factor_df = pd.read_csv("TH_LO_factors_D_20240129.csv", index_col="date")
factor_df
```

```python
# download data
ptt_df = yf.download(["PTT.BK"], interval="1d")

ptt_df["ret"] = np.log(ptt_df["Adj Close"] / ptt_df["Adj Close"].shift(1))
ptt_df.dropna(inplace=True)
ptt_df.index.name = "date"
ptt_df
```

```python
factor_df.join(ptt_df.ret)["ret"]
```

```python
combined_df = factor_df.merge(ptt_df, left_index=True, right_index=True)
combined_df
```
